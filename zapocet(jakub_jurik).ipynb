```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Nacteni dat z https://en.wikipedia.org/wiki/The_Economist_Democracy_Index
df = pd.read_html("edi.html")[5]
columns =  (['Region', '2024 rank', 'Country', 'Regime type']
           + list(range(2024, 2009, -1))) + [2008, 2006]
df.columns = columns
df.info()
```

```python
#1) Poradi podle indexu demokracie (predpokladam pro rok 2024)
df_ordered = pd.concat([df.iloc[:,0], df.iloc[:,2:]], axis=1)
df_ordered.set_index("Country", inplace=True)

df_ordered = df_ordered.sort_values(by=[2024], ascending=False)
df_ordered
```

```python
#2) Poradi podle poklesu indexu mezi 2010 a 2025
df_ordered2 = df_ordered
df_ordered2 = pd.concat([df_ordered.iloc[:,:17]], axis=1)
df_ordered2["Index diff"] = df_ordered2[2024] - df_ordered2[2010]
df_ordered2 = df_ordered2.sort_values(by=[2024], ascending=True)
df_ordered2
```

```python
# 3) Celkovy index pro EU (se zohlednenim poctu obyvatel)
pop = pd.read_html("population.html")[0]
pop = pop.iloc[:, 0:2]
pop.set_index("Location", inplace=True)

df3 = df
df3.set_index("Country", inplace=True)
df3 = pd.concat([df3.iloc[:,3:]], axis=1)
df3 = df3[df3.columns[::-1]]
df3 = df3.join(pop)
df3
```

```python
eu_states = [
    "Austria", "Belgium", "Bulgaria", "Croatia", "Cyprus",
    "Czech Republic", "Denmark", "Estonia", "Finland", "France",
    "Germany", "Greece", "Hungary", "Ireland", "Italy",
    "Latvia", "Lithuania", "Luxembourg", "Malta", "Netherlands",
    "Poland", "Portugal", "Romania", "Slovakia", "Slovenia",
    "Spain", "Sweden"
]

df_eu = df3.loc[eu_states]
df_eu
```

```python
pop_eu = df_eu.Population.sum()
pop_eu
```

```python
eu_index = df_eu.iloc[:,:-1].mul(df_eu.iloc[:, -1], axis=0).sum(axis=0) / pop_eu
eu_index # Indexy pro EU (se zohlednenim poctu obyvatel) pro jednotlive roky
```

```python
# 4) Provest klasifikaci zemi do skupin (dve az pet skupin) podle hodnot index(2024) resp. rozdíl (2024-2006) bez supervizoru + pocet obyvatel (logaritmus)
df3["Log(Pop)"] = np.log(df3["Population"])
df3["Diff(2006-2024)"] = df3[2024] - df3[2006]
df3
```

```python
df4 = df3[[2024, "Diff(2006-2024)", "Log(Pop)"]]
df4.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df4 = df4.dropna()
df4.columns = df4.columns.astype(str)

df_scaled = StandardScaler().fit_transform(df4)

kmeans = KMeans(n_clusters=4, random_state=42, n_init=20)
labels = kmeans.fit_predict(df_scaled)

df4["Groups"] = labels
df4 # Rozdeleny df do 4 skupin
```
